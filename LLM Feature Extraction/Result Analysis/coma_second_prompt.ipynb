# Coma — second prompt analysis

This notebook looks at the second iteration of the coma prompt. It loads the
newer coma extraction results and compares them, on the same notes, against the
earlier manual annotation sample, to see how the updated prompt changed the
coma labels.

```python
import pandas as pd
```

```python
path_result = "../outputs/coma_extraction_results_2026-05-13_22-55-31.csv"
df = pd.read_csv(path_result)
```

```python
path_result_old = "../Result Analysis/coma_manual_annotation_sample - coma_manual_annotation_sample.csv"
df_old_sample = pd.read_csv(path_result_old).sort_values(by='note_id').reset_index(drop=True)
```

```python
df_new_from_old = df.loc[df.note_id.isin(df_old_sample.note_id.tolist()), :].sort_values(by='note_id').reset_index(drop=True).copy()
```

```python
df_new_from_old[['note_id', 'llm_present']].equals(
    df_old_sample[['note_id', 'llm_present']]
)
```

```python
comparison = pd.concat(
   [
        df_new_from_old[['note_id', 'llm_present']].reset_index(drop=True),
        df_old_sample[['note_id', 'llm_present']].reset_index(drop=True)
    ],
    axis=1,
    keys=['new', 'old']
)

comparison[comparison['new']['llm_present'] != comparison['old']['llm_present']]
```

```python
df_null = df[
    (df['llm_present'].isna()) &
    (df['n_chunks'] >0 )
    
].head(10).copy()
```

```python
df_null[['note_id', 'hadm_id', 'subject_id','llm_present', 'llm_evidence_quote']].to_csv('coma_second_prompt_null.csv')
```
